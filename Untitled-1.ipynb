```python
# Cell 1: Imports
%matplotlib inline
import os
import numpy as np
from PIL import Image
from tqdm import tqdm
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
from collections import Counter
```

```python
poster_dir = r"C:\Users\91999\OneDrive\Desktop\movie\poster_downloads"
cache_file = "poster_features.npy"
cache_ids_file = "poster_ids.npy"

# Load pretrained model (VGG16 without top)
base_model = VGG16(weights="imagenet", include_top=False, pooling="avg")
```

```python
def extract_features(img_path, target_size=(224, 224)):
    try:
        img = image.load_img(img_path, target_size=target_size)
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        feat = base_model.predict(x, verbose=0)
        return feat.flatten()
    except Exception as e:
        print(f"Error processing {img_path}: {e}")
        return None
```

```python
poster_files = sorted([
    f for f in os.listdir(poster_dir)
    if f.lower().endswith(('.jpg', '.png'))
])
print(f"Found {len(poster_files)} poster images.")

if os.path.exists(cache_file) and os.path.exists(cache_ids_file):
    print("Loading cached features...")
    features = np.load(cache_file)
    valid_posters = np.load(cache_ids_file, allow_pickle=True).tolist()
else:
    features, valid_posters = [], []
    skipped = 0

    for poster_file in tqdm(poster_files, desc="Extracting features"):
        img_path = os.path.join(poster_dir, poster_file)
        feat = extract_features(img_path)
        if feat is not None:
            features.append(feat)
            valid_posters.append(poster_file)
        else:
            skipped += 1

    features = np.array(features)
    print(f"Extraction complete. {len(valid_posters)} posters processed, {skipped} skipped.")

    np.save(cache_file, features)
    np.save(cache_ids_file, np.array(valid_posters))

```

```python
def get_recommendations_with_scores(poster_index, top_n=5, min_similarity=0.0):
    sims = cosine_similarity([features[poster_index]], features)[0]
    indices = np.argsort(sims)[::-1][1:]  # exclude self
    recs = []
    for idx in indices:
        if len(recs) >= top_n:
            break
        if sims[idx] >= min_similarity:
            recs.append({'filename': valid_posters[idx], 'similarity': sims[idx]})
    return recs

def show_posters_grid_with_scores(recs, title="Recommended Posters"):
    n = len(recs)
    if n == 0:
        print("No recommendations found.")
        return
    cols = min(n, 5)
    rows = (n + cols - 1) // cols
    plt.figure(figsize=(3 * cols, 4 * rows))
    for i, rec in enumerate(recs):
        poster_file = rec['filename']
        sim = rec['similarity']
        img_path = os.path.join(poster_dir, poster_file)
        plt.subplot(rows, cols, i + 1)
        img = Image.open(img_path)
        plt.imshow(img)
        plt.title(f"{poster_file}\nSim: {sim:.2f}")
        plt.axis('off')
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

def recommend_and_show(poster_index=None, filename=None, top_n=5, min_similarity=0.0):
    if filename is not None:
        if filename not in valid_posters:
            print(f"File {filename} not found.")
            return
        poster_index = valid_posters.index(filename)
    if poster_index is None or poster_index >= len(valid_posters):
        print("Invalid poster index.")
        return
    print(f"Recommendations for {valid_posters[poster_index]} (min similarity {min_similarity}):")
    recs = get_recommendations_with_scores(poster_index, top_n, min_similarity)
    print(recs)
    show_posters_grid_with_scores(
        [{'filename': valid_posters[poster_index], 'similarity': 1.0}],
        title="Query Poster"
    )
    show_posters_grid_with_scores(recs, title="Recommended Posters")
```
